Görselleştirme, verileri farklı grafikler ve çizimler kullanarak görselleştirmek için kullanılan bir tekniktir. Veri biliminde, veri kümesini anlamak ve veriler arasındaki ilişkiyi bulmak için genellikle veri görselleştirme tekniklerini kullanırız. Görselleştirme ayrıca veri kümesinde daha fazla analiz için kullanılan örüntüyü bulmaya da yardımcı olabilir.

Python'da Veri Görselleştirme için kullanılan Matplotlib, Seaborn, Plotly gibi farklı teknikler / kütüphaneler vardır. Ancak tüm bu kütüphaneleri kullanırken görselleştirmek istediğimiz grafik türünü ve görselleştirmek için ihtiyaç duyduğumuz argümanları tanımlamamız gerekir.

**Bu makalede, tek bir kod satırında tüm Veri Görselleştirme sürecini otomatikleştirebilen bir python kütüphanesi AutoViz hakkında bilgi edineceğiz.**

AutoViz, sadece bir satır kod ile herhangi bir veri kümesinin otomatik görselleştirmesini gerçekleştirir. AutoViz en önemli özellikleri bulabilir ve yalnızca otomatik olarak seçilen özellikleri kullanarak etkili görselleştirmeler çizebilir. Ayrıca, AutoViz inanılmaz derecede hızlıdır, bu nedenle saniyeler içinde görselleştirme oluşturur.

# **2. Advertising Veri Kümesi**

Burada, AutoViz'in görselleştirmeyi buna göre oluşturması için bağımlı değişken olan depVar argümanını daha ileteceğiz. Bu veri kümesinde, bağımlı değişkenin 'sales' olduğunu biliyoruz.

```python
df = AV.AutoViz('advertising.csv',
                depVar='Sales')
```

**a. Satışlara Karşı Değişkenlerin Dağılım Grafiği**

5.resim.jpg

'Satışlar' bağımlıdır.

Benzer şekilde, 'Satışları' bağımlı değişken olarak tutarak histogram, keman grafiği, ısı haritası vb. oluşturacağını göreceksiniz.

AutoViz'i çağırırken geçebileceğimiz diğer bazı argümanlar şunlardır:

a. sep, verilerin ayrıldığı ayırıcıdır, varsayılan olarak ',' şeklindedir.

b. target, veri kümesindeki hedef değişkendir.

c. chart_format, görüntülenen grafiğin biçimidir.

d. max_row_analyzed, analiz edilecek satır sayısını tanımlamak için kullanılır.

e. max_cols_analyzed, analiz edilecek sütun sayısını tanımlamak için kullanılır.

Bu yazıda veri kümelerini sadece bir satır kod ile görselleştirebileceğimizi ve buna göre veri kümesindeki örüntüleri bulabileceğimizi gördük.
